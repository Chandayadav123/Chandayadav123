```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices.csv')
df
```

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old



```python
reg.predict([[3000, 3, 40]])
```

**Machine Learning With Python: Linear Regression Multiple Variables** , Hiring CSV

```python
df = pd.read_csv('hiring.csv')
df
```

```python
df.experience.mode()
```

```python
df.experience = df.experience.fillna(df.experience.mode())
df
```

```python
df = df.rename(columns={'test_score(out of 10)': 'test_score'})
df = df.rename(columns={'interview_score(out of 10)': 'interview_score','salary($)': 'salary'})
```

```python
text_to_int = {'eleven': 11, 'seven': 7, 'five': 5, 'two': 2, 'ten':10, 'three': 3}
def convert_experience(text):
  try:
    return int(text)  # Try converting to integer first
  except ValueError:
    return text_to_int.get(text)
```

```python
df['experience'] = df['experience'].apply(convert_experience)
```

```python
print(df)
```

```python
df.test_score = df.test_score.fillna(df.test_score.mode())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('salary',axis='columns'),df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2, 9, 6]])
```

```python
reg.predict([[12, 10, 10]])
```
